```python
def addall(x, *args):
    print(x, args)
    res = x 
    for y in args:
        res += y
    return(res)
addall(1,2,3,4,5)
```

```python
def make_two_lists( **kwargs):
    print(kwargs)
    key, value = [],[]
    for k,v in kwargs.items():
        key.append(k)
        value.append(v)
    return([key,value])

make_two_lists(david= 'M', Mary= 'F', John= 'M')
```
